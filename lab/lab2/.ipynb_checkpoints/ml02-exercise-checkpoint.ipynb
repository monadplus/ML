```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
 
#loading
(train_X, train_y), (test_X, test_y) = mnist.load_data()
 
#shape of dataset
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
 
#plotting
for i in range(9):  
    plt.subplot(330 + 1 + i)
    plt.imshow(train_X[i], cmap=plt.cm.gray)
    plt.axis('off')
```

## Image Compression (truncated SVD)

```python
X = train_X[0]
print(X.shape)
plt.imshow(X, cmap=plt.cm.gray)
plt.axis('off')
```

```python
from numpy.linalg import svd

U,σ,Vt = svd(X)
Σ = np.diag(σ)

np.allclose(X, U @ Σ @ Vt)
```

```python
m,n = X.shape
j = 0
plt.figure(figsize=(14, 6))
for k in [1 << i for i in range(10)]:
    plt.subplot(2,5,j+1)
    j += 1
    X_approx = U[:,:k] @ Σ[:k,:k] @ Vt[:k,:]
    plt.imshow(X_approx, cmap=plt.cm.gray)
    plt.axis('off')
    plt.title('k = ' + str(k))
```

### Singular Values

```python
# Notice σ_1 >= σ_2 >= σ_r
plt.figure(figsize=(16,4))
plt.subplot(131)
plt.plot(σ)
plt.title('Singular values')

# plt.subplot(132)
# plt.semilogy(σ)
# plt.title('Singular values in log scale')

plt.subplot(132)
plt.plot(np.cumsum(σ) / np.sum(σ))
plt.title('Cumulative sum of singular values')
```

## 'eigendigits'

In this last section we are going to apply SVD to a collection of images, each image will be a "row" in a matrix and we will perform SVD on this matrix.

```python
h = train_X.shape[1] # 28
w = train_X.shape[2] # 28
X = np.zeros((train_X.shape[0], h*w))

for i, image in enumerate(train_X):
    X[i,:] = image.reshape(1 , -1) # flatten
    
X.shape
```

Print data

```python
n=10
plt.figure(figsize=(n, n))
for i in range(n*n):
    plt.subplot(n,n,i+1)
    idx = np.random.randint(n)
    plt.imshow(X[i,:].reshape(h,w), cmap=plt.cm.gray)
    plt.axis('off')
```

Compute mean digit; subtract mean digit from all our digits.

```python
mean_digit = X.mean(axis=0)
plt.imshow(np.reshape(mean_digit, (h,w)), cmap=plt.cm.gray)
plt.title("Mean Digit")
plt.axis('off')
plt.show()
```

```python
# Center data
cX = X - mean_digit
```

Redraw centered digits

```python
n=10
plt.figure(figsize=(n, n))
for i in range(n*n):
    plt.subplot(n,n,i+1)
    idx = np.random.randint(n)
    plt.imshow(cX[i,:].reshape(h,w), cmap=plt.cm.gray)
    plt.axis('off')
```

```python
%%time
U, d, Vt = svd(cX, full_matrices=False)
```

```python
U.shape, d.shape, Vt.shape
```

```python
np.allclose(cX, U @ np.diag(d) @ Vt)
```

Show first eigvectors

```python
n=10
plt.figure(figsize=(n, n))
for i in range(n*n):
    plt.subplot(n,n,i+1)
    idx = np.random.randint(n)
    plt.imshow(Vt[i,:].reshape(h,w), cmap=plt.cm.gray)
    plt.axis('off')
```

Plot singular values

```python
plt.figure(figsize=(16,4))
plt.subplot(131)
plt.plot(d)
plt.title('Singular values')

plt.subplot(132)
plt.plot(np.cumsum(d) / np.sum(d))
plt.title('Cumulative sum of singular values')
```

### Columns of $V$ as new basis for images

Remember that the SVD decomposition of our image matrix is

$$
X = \sigma_1\mathbf{u}_1\mathbf{v}_1^T + \sigma_2\mathbf{u}_2\mathbf{v}_2^T + ... + \sigma_n\mathbf{u}_n\mathbf{v}_n^T
$$

It is useful to see what happens with a single row of $X$ (that is, one image). If you apply the decomposition you can see that it is now expressed as a **linear combination** of rows of $V^T$, i.e. columns of $V$; the coefficients of the linear combination are given by the corresponding component of the columns of $U$ multiplied by the corresponding singular value (see below). So, columns of $V$ represent new orthonormal basis for rows of $X$. Each **eigendigit** is thus an element of the new orthonormal basis for images (rows) of $X$.

$$
\begin{aligned}
X[i,:] = \mathbf{x}_i &= \sigma_1 u[i]_1 \mathbf{v}_1^T + \sigma_2 u[i]_2 \mathbf{v}_2^T + ... + \sigma_n u[i]_n \mathbf{v}_n^T \\
    &= \alpha_1 \mathbf{v}_1^T + \alpha_2 \mathbf{v}_2^T + ... + \alpha_n \mathbf{v}_n^T 
\end{aligned}
$$

where each $\alpha_j = \sigma_j u[i]_j$; the coefficient vector $\mathbf{\alpha}$ constitutes the new "fingerprint" of the image $\mathbf{x}_i$. Given the space of eigenvectors in $V$, it is enough to reconstruct an image to give its *signature* coefficients $\alpha$.

Again, to save space, we can only keep the first terms in the sum (say the first $k$ terms corresponding the most important eigenvectors) and so with this truncated sum we will only  *approximate*  rows of $X$. So, in order to store the whole collection of images, it is enough to store the most important eigenfaces, and then for each image we just need to store the coefficients used to restore it with satisfactory precision.

In essence, by truncating the sum we are performing **dimensionality reduction** from this high dimensional space (each image has $28 \times 28 = 784$ features -- namely its pixels) down to $k$ coefficients.

### Reconstruct digit from few eigendigits

```python
i = np.random.randint(n) # select random image
alpha = U[i,:] @ np.diag(d)
print(alpha.shape)
plt.imshow((mean_digit + cX[i,:]).reshape(h,w), cmap=plt.cm.gray)
plt.axis('off')
```

```python
plt.figure(figsize=(15, 40))
j = 1
k_range = [5, 50, 100, 300, 500, Vt.shape[0]] 
for k in k_range:
    plt.subplot(1,len(k_range),j)
    j += 1
    reconstructed = mean_digit + alpha[:k] @ Vt[:k,:]
    plt.imshow(reconstructed.reshape(h,w), cmap=plt.cm.gray)
    plt.axis('off')
    plt.title('k = ' + str(k))
```

### "Folding in" new images

We can project an image $\mathbf{x}$ _outside_ our matrix $X$ onto the same space and compute its coefficients using the transformation $$\mathbf{\alpha}' = \mathbf{x}'  V$$ 

This way we can represent
any vector as the linear combination of eigendigits; using the same trick as above we can reconstruct it using the truncated space

```python
import random

# pick an image outside X
image = random.choice(test_X)
new_x = image.reshape(1 , -1) - mean_digit
new_alpha = new_x @ Vt.T
new_alpha = new_alpha.reshape(-1)
plt.imshow(image, cmap=plt.cm.gray)
```

#### Reconstruct the digit using a few eigendigits

```python
plt.figure(figsize=(15, 40))
j = 1
k_range = [5, 50, 100, 300, 500, Vt.shape[0]] 
for k in k_range:
    plt.subplot(1,len(k_range),j)
    j += 1
    reconstructed = mean_digit + new_alpha[:k] @ Vt[:k,:]
    plt.imshow(reconstructed.reshape(h,w), cmap=plt.cm.gray)
    plt.axis('off')
    plt.title('k = ' + str(k))
```

Notice, we started from the eigendigits of our training dataset and we could generate the test image.

## 2D visualization of dataset

```python
ones = [image for i, image in enumerate(train_X) if train_y[i] == 1]
fives = [image for i, image in enumerate(train_X) if train_y[i] == 5]

len(ones)
len(fives)
```

```python
# get coordinates for both using 2 eigendigits
eigen_modes = [0,1]
def get_coord(images):
    res = np.zeros((len(images), 2))
    for image in images:
        x = image.reshape(1 , -1) - mean_digit
        alpha = (x @ Vt.T).reshape(-1)
        res[i,:] = alpha[eigen_modes]
    return res
ones_coord = get_coord(ones)
fives_coord = get_coord(fives)
```

```python
# plot them in 2D
plt.plot(ones_coord[:,0], ones_coord[:,1], '.', color='m', label='ones')
plt.plot(fives_coord[:,0], fives_coord[:,1], 'x', color='b', label='fives')
plt.legend()
```
